```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('Dataset .csv')
df.head()
```

```python
df.drop([
    'Restaurant ID', 'Restaurant Name', 'Country Code', 'City',
    'Address', 'Locality', 'Locality Verbose',
    'Longitude', 'Latitude', 'Currency',
    'Switch to order menu',
    'Rating color', 'Rating text'
], axis=1, inplace=True)
```

```python
df.dropna(subset=['Cuisines'], inplace=True)
df.isnull().sum()
```

```python
le = LabelEncoder()

df['Cuisines'] = le.fit_transform(df['Cuisines'])
df['Has Table booking'] = le.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = le.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = le.fit_transform(df['Is delivering now'])
```

```python
features = df[['Cuisines', 'Average Cost for two',
               'Has Online delivery', 'Has Table booking']]
```

```python
similarity_matrix = cosine_similarity(features)
```

```python
def recommend_restaurants(user_index, top_n=5):
    similarity_scores = list(enumerate(similarity_matrix[user_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    recommended_indices = [i[0] for i in similarity_scores[1:top_n+1]]
    return df.iloc[recommended_indices]
```

```python
recommended = recommend_restaurants(user_index=10, top_n=5)
recommended
```
